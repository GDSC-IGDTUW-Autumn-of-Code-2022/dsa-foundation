```python
from math import sqrt
def euclidean(row1,row2):
    dist=0.0
    for i in range(len(row1)-1,):
      dist+=(row1[i]-row2[i])**2
    return sqrt(dist)  
    


import numpy as np
def get_neighbors(x,row,l):
	dist=[]
	for i in x:
		distance = euclidean(row, i)
		dist.append((row, distance))
	dist.sort(key=lambda tup: tup[1])
	neighbors = []
	for j in range(l):
		neighbors.append(dist[j][0])
	return neighbors


from scipy.stats import mode
def predict(train, test_row, num_neighbors):
	neighbors = get_neighbors(train, test_row, num_neighbors)
	output_values = [row[-1] for row in neighbors]
	prediction = max(set(output_values), key=output_values.count)
	return prediction
 

from sklearn.datasets import load_iris
iris_data=load_iris()
def iris_predict(iris_data,test_row,m):
    train=[]
    for i in range(len(iris_data.data)):
        train.append(list(iris_data.data[i])+[iris_data.target[i]])
    prediction = predict(train,test_row,m)
    return prediction, iris_data.target_names[int(prediction)]
    
iris_predict(iris_data,[4.9, 3. , 1.4, 0.2],3)
```
